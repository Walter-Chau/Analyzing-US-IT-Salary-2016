```python
#Have a look at the distribution first
ax = df['Annual_Base_Pay']\
.plot(kind='box', title='Distribution of Annual Base Pay')
ax.set_xlabel('Categories')
ax.set_ylabel('Count')
```

```python
#The distribution is significantly imbalanced, we need to remove the outliers
df['Annual_Base_Pay'].quantile(0.25)
df['Annual_Base_Pay'].quantile(0.75)
Q1 = df['Annual_Base_Pay'].quantile(0.25)
Q3 = df['Annual_Base_Pay'].quantile(0.75)
IQR = Q3 - Q1
IQR
```

```python
lower_lim = Q1 - 1.5*IQR
upper_lim = Q3 + 1.5*IQR
```

```python
outliers_15_low = (df['Annual_Base_Pay'] < lower_lim)
outliers_15_up = (df['Annual_Base_Pay'] > upper_lim)
```

```python
df = df[~(outliers_15_low | outliers_15_up)]
df.reset_index(drop=True)
```

```python
#The distribution is significantly clearer than before
ax = df['Annual_Base_Pay']\
.plot(kind='box', title='Distribution of Annual Base Pay')
ax.set_xlabel('Categories')
ax.set_ylabel('Count')
```

```python
#What are the locations with the highest annual base pay (minimum of 10)?
df['Location_State'].value_counts()
```

```python
#The answer is CA
df.groupby('Location_State')['Annual_Base_Pay']\
.agg(['mean','count'])\
.query('count >= 15')\
.sort_values('mean')['mean']\
.plot(kind='barh', figsize=(12, 5), title = 'Average annual pay by Location')

ax.set_xlabel('Average annual pay')
plt.show()
```
